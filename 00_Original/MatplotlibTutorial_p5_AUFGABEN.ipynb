1. Aufgabe:

Erzeuge einen Plot der Funktion 
$$z = sin(x^3) + cos(y^2)$$

````python
# progr4book

import numpy as np
import matplotlib.pyplot as plt

def sin2d(x, y):
    return np.sin(x**3) + np.cos(y**2)

X, Y = np.meshgrid(np.linspace(0, 5*np.pi, 200), 
                   np.linspace(0, 5*np.pi, 200))
Z = sin2d(X, Y) 

plt.imshow(Z, origin='lower')
plt.show()
````

2. Aufgabe:

Bei dieser Aufgabe handelt es sich um eine herzige Angelegenheit. Einfach fr die folgende Funktion einen Konturplott erzeugen, und es wird sich zeigen warum:

$$ x^2 + (y - (x^2)^{1 \over 5}) ^ 2$$

````python
# prog4book
import matplotlib.pyplot as plt
import numpy as np

y, x = np.ogrid[-1:2:100j, -1:1:100j]


plt.contour(x.ravel(), 
            y.ravel(), 
            x ** 2 + (y - ((x ** 2) ** (1.0 / 5))) ** 2, 
            [1],
            colors='red')
plt.axis('equal')
plt.show()
````
