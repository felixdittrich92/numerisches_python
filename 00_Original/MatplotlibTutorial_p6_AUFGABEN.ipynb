    Bei der Bundestagswahl in Deutschland gab es im Jahre 2017 folgende prozentuale Verteilungen f체r die einzelnen Parteien:
    
    CDU/CSU: 32,9, SPD: 20,5, AfD: 12,6, FDP: 10,7, Linke: 9,2, Gr체ne: 8,9, Sonstige: 5

Erzeuge ein S채ulendiagramm mit diesen Daten.
